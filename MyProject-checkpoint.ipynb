```python
import tkinter as tk
from joblib import load
import pandas as pd

# Load the trained model from the file
model_filename = "random_forest_model.joblib"
clf = load(model_filename)

# Function to predict the result and provide recommendations
def predict_result():
    # Get the input values from the text boxes
    age = float(entry_age.get())
    study = float(entry_study.get())
    att = float(entry_att.get())
    teach = float(entry_teach.get())
    math = float(entry_math.get())
    phyc = float(entry_phyc.get())
    life = float(entry_life.get())
    geo = float(entry_geo.get())
    eng = float(entry_eng.get())

    # Create a dictionary to store the input values
    student_data = {
        'Age': age,
        'study time': study,
        'Attendance': att,
        'Teacher': teach,
        'Mathematics': math,
        'Physical Sciences': phyc,
        'Life Sciences': life,
        'Geography': geo,
        'English': eng
    }

    # Convert the dictionary to a pandas DataFrame
    student_df = pd.DataFrame([student_data])

    # Use the loaded model to make a prediction
    prediction = clf.predict(student_df)[0]

    # Determine the prediction result based on the model's output
    if prediction == 0:
        result_text = "The student is likely to fail."
        recommendation_text = "Don't be discouraged. Failure is a stepping stone to success. Keep working hard."
    else:
        result_text = "The student is likely to pass."
        recommendation_text = "Congratulations! Keep up the good work."

    # Display the prediction result and recommendation in the respective text boxes
    result_text_box.delete(1.0, tk.END)  # Clear any previous content
    result_text_box.insert(tk.END, result_text)  # Insert the prediction result

    recommendation_text_box.delete(1.0, tk.END)  # Clear any previous content
    recommendation_text_box.insert(tk.END, recommendation_text)  # Insert the general recommendation

    # Provide subject-specific recommendations
    subjects = ["Mathematics", "Physical Sciences", "Life Sciences", "Geography", "English"]
    scores = [math, phyc, life, geo, eng]  # Use the input scores for each subject

    for subject, score in zip(subjects, scores):
        recommendation_text_box.insert(tk.END, f"\n\nSubject: {subject}\n")
        if score >= 50:
            recommendation_text_box.insert(tk.END, f"\nYou passed {subject}, but there's always room for improvement.")
            # Provide specific recommendations based on subject and score
            if subject == "Mathematics":
                recommendation_text_box.insert(tk.END, "\nConsider practicing more challenging problems to strengthen your understanding.")
            elif subject == "Physical Sciences":
                recommendation_text_box.insert(tk.END, "\nTry to understand the concepts in depth, and practice problem-solving techniques.")
            elif subject == "Life Sciences":
                recommendation_text_box.insert(tk.END, "\nReview the topics you find difficult and seek help from teachers or classmates.")
            elif subject == "Geography":
                recommendation_text_box.insert(tk.END, "\nPay attention to the details and try to remember the key points.")
            elif subject == "English":
                recommendation_text_box.insert(tk.END, "\nPractice reading and writing regularly to improve your skills.")
        else:
            recommendation_text_box.insert(tk.END, f"\nYou failed {subject}. Here are some recommendations to improve:")
            # Provide specific recommendations based on subject and score
            if subject == "Mathematics":
                recommendation_text_box.insert(tk.END, "\nFocus on understanding core concepts before attempting practice problems.")
            elif subject == "Physical Sciences":
                recommendation_text_box.insert(tk.END, "\nReview the topics you find difficult and seek help from teachers or classmates.")
            elif subject == "Life Sciences":
                recommendation_text_box.insert(tk.END, "\nPractice more and try to understand the concepts in depth.")
            elif subject == "Geography":
                recommendation_text_box.insert(tk.END, "\nPay attention to the details and try to remember the key points.")
            elif subject == "English":
                recommendation_text_box.insert(tk.END, "\nPractice reading and writing regularly to improve your skills.")

# Function to clear the form
def clear_form():
    # Clear the text boxes
    entry_age.delete(0, tk.END)
    entry_study.delete(0, tk.END)
    entry_att.delete(0, tk.END)
    entry_teach.delete(0, tk.END)
    entry_math.delete(0, tk.END)
    entry_phyc.delete(0, tk.END)
    entry_life.delete(0, tk.END)
    entry_geo.delete(0, tk.END)
    entry_eng.delete(0, tk.END)

    # Clear the text boxes
    result_text_box.delete(1.0, tk.END)
    recommendation_text_box.delete(1.0, tk.END)

# Create the main window
root = tk.Tk()
root.title("Student Result Prediction")

# Create the main frame
main_frame = tk.Frame(root)
main_frame.pack(fill=tk.BOTH, expand=True)

# Create heading with larger font size
heading_label = tk.Label(main_frame, text="Student Result Prediction", font=("Arial", 24, "bold"))
heading_label.pack(pady=30)

# Create a frame for input fields and prediction/recommendation results
input_result_frame = tk.Frame(main_frame)
input_result_frame.pack(pady=20, fill=tk.BOTH, expand=True)

# Create input fields with labels aligned to the left
input_frame = tk.Frame(input_result_frame)
input_frame.pack(side=tk.LEFT, padx=20)

label_age = tk.Label(input_frame, text="Age:", font=("Arial", 14))
label_age.grid(row=0, column=0, padx=10, pady=10, sticky=tk.W)
entry_age = tk.Entry(input_frame, font=("Arial", 14))
entry_age.grid(row=0, column=1, padx=10, pady=10)

label_study = tk.Label(input_frame, text="Study Time:", font=("Arial", 14))
label_study.grid(row=1, column=0, padx=10, pady=10, sticky=tk.W)
entry_study = tk.Entry(input_frame, font=("Arial", 14))
entry_study.grid(row=1, column=1, padx=10, pady=10)

label_att = tk.Label(input_frame, text="Attendance:", font=("Arial", 14))
label_att.grid(row=2, column=0, padx=10, pady=10, sticky=tk.W)
entry_att = tk.Entry(input_frame, font=("Arial", 14))
entry_att.grid(row=2, column=1, padx=10, pady=10)

label_teach = tk.Label(input_frame, text="Teacher:", font=("Arial", 14))
label_teach.grid(row=3, column=0, padx=10, pady=10, sticky=tk.W)
entry_teach = tk.Entry(input_frame, font=("Arial", 14))
entry_teach.grid(row=3, column=1, padx=10, pady=10)

label_math = tk.Label(input_frame, text="Mathematics:", font=("Arial", 14))
label_math.grid(row=4, column=0, padx=10, pady=10, sticky=tk.W)
entry_math = tk.Entry(input_frame, font=("Arial", 14))
entry_math.grid(row=4, column=1, padx=10, pady=10)

label_phyc = tk.Label(input_frame, text="Physical Sciences:", font=("Arial", 14))
label_phyc.grid(row=5, column=0, padx=10, pady=10, sticky=tk.W)
entry_phyc = tk.Entry(input_frame, font=("Arial", 14))
entry_phyc.grid(row=5, column=1, padx=10, pady=10)

label_life = tk.Label(input_frame, text="Life Sciences:", font=("Arial", 14))
label_life.grid(row=6, column=0, padx=10, pady=10, sticky=tk.W)
entry_life = tk.Entry(input_frame, font=("Arial", 14))
entry_life.grid(row=6, column=1, padx=10, pady=10)

label_geo = tk.Label(input_frame, text="Geography:", font=("Arial", 14))
label_geo.grid(row=7, column=0, padx=10, pady=10, sticky=tk.W)
entry_geo = tk.Entry(input_frame, font=("Arial", 14))
entry_geo.grid(row=7, column=1, padx=10, pady=10)

label_eng = tk.Label(input_frame, text="English:", font=("Arial", 14))
label_eng.grid(row=8, column=0, padx=10, pady=10, sticky=tk.W)
entry_eng = tk.Entry(input_frame, font=("Arial", 14))
entry_eng.grid(row=8, column=1, padx=10, pady=10)

# Create a frame for prediction and recommendation results aligned to the right
result_frame = tk.Frame(input_result_frame)
result_frame.pack(side=tk.RIGHT, padx=20)

# Prediction Results section
result_label = tk.Label(result_frame, text="Prediction Results", font=("Arial", 16, "bold"))
result_label.pack(pady=10)

result_text_box = tk.Text(result_frame, height=5, width=40, font=("Arial", 14))
result_text_box.pack(pady=10)

# Recommendations section
recommendation_label = tk.Label(result_frame, text="Recommendations", font=("Arial", 16, "bold"))
recommendation_label.pack(pady=10)

recommendation_text_box = tk.Text(result_frame, height=15, width=40, font=("Arial", 14))
recommendation_text_box.pack(pady=10)

# Create the buttons
button_frame = tk.Frame(main_frame)
button_frame.pack(pady=20)

clear_button = tk.Button(button_frame, text="Clear Form", command=clear_form, width=20, font=("Arial", 14))
clear_button.pack(side=tk.LEFT, padx=10)

predict_button = tk.Button(button_frame, text="Predict", command=predict_result, width=20, font=("Arial", 14))
predict_button.pack(side=tk.RIGHT, padx=10)

# Start the main event loop
root.mainloop()
```
